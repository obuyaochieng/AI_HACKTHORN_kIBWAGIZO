# CELL 1: TRAIN AND SAVE RAINFALL ZONE MODEL

```python
# ============================================================================
# CELL 1: SETUP AND DATA LOADING
# ============================================================================

import pandas as pd
import numpy as np
import pickle
import json
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import warnings
warnings.filterwarnings('ignore')

print("üåßÔ∏è RAINFALL ZONE CLASSIFICATION SYSTEM")
print("=" * 80)
print("Training rainfall zone model for farm insurance...")

# Load your rainfall data
# REPLACE THIS WITH YOUR ACTUAL FILE PATH
try:
    df = pd.read_csv('your_rainfall_data.csv')  # Update this path
    print(f"‚úÖ Data loaded successfully from CSV!")
except FileNotFoundError:
    print("‚ö†Ô∏è  File not found. Creating sample data for demonstration...")
    # Create realistic sample data
    np.random.seed(42)
    n_farms = 190
    years = list(range(2018, 2025))
    
    data = []
    for farm_id in range(1, n_farms + 1):
        for year in years:
            for month in range(1, 13):
                # Create different rainfall patterns
                if farm_id <= 30:  # High rainfall zone
                    base = 100
                    variability = 0.3
                elif farm_id <= 80:  # Medium rainfall zone
                    base = 50
                    variability = 0.5
                elif farm_id <= 130:  # Low rainfall zone
                    base = 20
                    variability = 0.7
                else:  # Arid zone
                    base = 5
                    variability = 0.9
                
                # Add seasonal pattern (peaks in Nov-Dec)
                seasonal = 1 + 0.8 * np.sin(2 * np.pi * (month - 11) / 12)
                
                # Generate rainfall
                rain = np.random.gamma(shape=2, scale=base * seasonal)
                rain *= (1 + np.random.uniform(-variability, variability))
                
                # Add dry days (more in arid zones)
                dry_prob = 0.7 if farm_id > 130 else 0.5
                if np.random.random() < dry_prob:
                    rain = 0
                
                data.append({
                    'id_number': f'farm_{farm_id:04d}',
                    'year': year,
                    'month': month,
                    'rainfall_mm': max(0, rain)
                })
    
    df = pd.DataFrame(data)
    print(f"‚úÖ Created sample data: {n_farms} farms, {len(years)} years")

print(f"\nüìä DATA OVERVIEW:")
print(f"   ‚Ä¢ Shape: {df.shape}")
print(f"   ‚Ä¢ Unique farms: {df['id_number'].nunique()}")
print(f"   ‚Ä¢ Date range: {df['year'].min()} to {df['year'].max()}")
print(f"   ‚Ä¢ Total records: {len(df):,}")
print(f"\nüìã Sample data (first 5 rows):")
print(df.head())
```

# FEATURE ENGINEERING FOR ZONES

```python
# ============================================================================
# CELL 2: FEATURE ENGINEERING
# ============================================================================

print("\n" + "=" * 80)
print("üîß FEATURE ENGINEERING")
print("=" * 80)

print("Creating features for each farm...")

def create_farm_features(df):
    """Create comprehensive features for each farm"""
    features_list = []
    
    for farm_id in df['id_number'].unique():
        farm_data = df[df['id_number'] == farm_id]
        
        # Initialize feature dictionary
        features = {'farm_id': farm_id}
        
        # 1. BASIC STATISTICS
        features['total_rainfall'] = float(farm_data['rainfall_mm'].sum())
        features['mean_rainfall'] = float(farm_data['rainfall_mm'].mean())
        features['median_rainfall'] = float(farm_data['rainfall_mm'].median())
        features['std_rainfall'] = float(farm_data['rainfall_mm'].std())
        
        # Avoid division by zero
        if features['mean_rainfall'] > 0:
            features['cv_rainfall'] = float(features['std_rainfall'] / features['mean_rainfall'])
        else:
            features['cv_rainfall'] = 0.0
        
        # 2. RAINY DAYS ANALYSIS
        rainy_days = farm_data[farm_data['rainfall_mm'] > 0]
        features['rainy_days_count'] = int(len(rainy_days))
        features['rainy_days_percent'] = float((len(rainy_days) / len(farm_data)) * 100)
        
        if len(rainy_days) > 0:
            features['mean_rain_intensity'] = float(rainy_days['rainfall_mm'].mean())
        else:
            features['mean_rain_intensity'] = 0.0
        
        # 3. DROUGHT ANALYSIS (Dry spells)
        dry_spells = []
        current_spell = 0
        
        for val in farm_data['rainfall_mm'].values:
            if val == 0:
                current_spell += 1
            else:
                if current_spell > 0:
                    dry_spells.append(current_spell)
                current_spell = 0
        
        if current_spell > 0:
            dry_spells.append(current_spell)
        
        if dry_spells:
            features['max_dry_spell'] = int(max(dry_spells))
            features['avg_dry_spell'] = float(np.mean(dry_spells))
            features['dry_spell_count'] = int(len(dry_spells))
        else:
            features['max_dry_spell'] = 0
            features['avg_dry_spell'] = 0.0
            features['dry_spell_count'] = 0
        
        # 4. SEASONALITY ANALYSIS
        monthly_means = farm_data.groupby('month')['rainfall_mm'].mean()
        if len(monthly_means) > 0:
            mean_monthly = monthly_means.mean()
            if mean_monthly > 0:
                features['seasonality_index'] = float(monthly_means.std() / mean_monthly)
            else:
                features['seasonality_index'] = 0.0
            
            # Wet months ratio
            wet_months = len(monthly_means[monthly_means > mean_monthly])
            features['wet_months_ratio'] = float(wet_months / 12)
        else:
            features['seasonality_index'] = 0.0
            features['wet_months_ratio'] = 0.0
        
        # 5. EXTREME EVENTS
        if len(farm_data) > 0:
            p95 = farm_data['rainfall_mm'].quantile(0.95)
            p99 = farm_data['rainfall_mm'].quantile(0.99)
            features['p95_events'] = int(len(farm_data[farm_data['rainfall_mm'] > p95]))
            features['p99_events'] = int(len(farm_data[farm_data['rainfall_mm'] > p99]))
        else:
            features['p95_events'] = 0
            features['p99_events'] = 0
        
        # 6. TREND ANALYSIS
        yearly_totals = farm_data.groupby('year')['rainfall_mm'].sum()
        if len(yearly_totals) > 1:
            x = np.array(yearly_totals.index)
            y = yearly_totals.values
            z = np.polyfit(x, y, 1)  # Linear trend
            features['trend_slope'] = float(z[0])
            
            if np.mean(y) > 0:
                features['trend_percent'] = float((z[0] * len(x)) / np.mean(y) * 100)
            else:
                features['trend_percent'] = 0.0
        else:
            features['trend_slope'] = 0.0
            features['trend_percent'] = 0.0
        
        # 7. MONTHLY PATTERNS
        for month in range(1, 13):
            month_data = farm_data[farm_data['month'] == month]['rainfall_mm']
            if len(month_data) > 0:
                features[f'month_{month}_mean'] = float(month_data.mean())
                features[f'month_{month}_total'] = float(month_data.sum())
            else:
                features[f'month_{month}_mean'] = 0.0
                features[f'month_{month}_total'] = 0.0
        
        features_list.append(features)
    
    # Create DataFrame
    features_df = pd.DataFrame(features_list)
    features_df = features_df.set_index('farm_id')
    
    print(f"‚úÖ Created {len(features_df.columns)} features for {len(features_df)} farms")
    return features_df

# Create features
features_df = create_farm_features(df)

print(f"\nüìà FEATURE SUMMARY:")
print(f"   ‚Ä¢ Total features: {len(features_df.columns)}")
print(f"   ‚Ä¢ Farms analyzed: {len(features_df)}")
print(f"\nüìä Key statistics:")
key_stats = features_df[['total_rainfall', 'rainy_days_percent', 'max_dry_spell', 'cv_rainfall']].describe()
print(key_stats.round(2))
```

```python
 # ============================================================================
# CELL 3: FEATURE SELECTION
# ============================================================================

print("\n" + "=" * 80)
print("üéØ FEATURE SELECTION")
print("=" * 80)

print("Selecting most important features for zone classification...")

# Based on domain knowledge and your analysis
key_features = [
    # Core rainfall metrics
    'total_rainfall', 'mean_rainfall', 'cv_rainfall',
    
    # Rainfall patterns
    'rainy_days_percent', 'mean_rain_intensity',
    
    # Drought indicators
    'max_dry_spell', 'avg_dry_spell',
    
    # Seasonality
    'seasonality_index', 'wet_months_ratio',
    
    # Extreme events
    'p95_events', 'p99_events',
    
    # Trends
    'trend_slope', 'trend_percent',
    
    # Critical months (from your analysis)
    'month_11_mean',  # November (wettest)
    'month_12_mean',  # December
    'month_6_mean',   # June (driest)
    'month_3_mean',   # March
    'month_4_mean'    # April
]

# Filter to available features
available_features = []
for feature in key_features:
    if feature in features_df.columns:
        available_features.append(feature)
        print(f"  ‚úì {feature}")
    else:
        print(f"  ‚úó {feature} (not available)")

print(f"\n‚úÖ Selected {len(available_features)} features")

# Create feature matrix
X = features_df[available_features].copy()

# Handle any missing values
X = X.fillna(X.mean())

print(f"\nüìä Feature matrix shape: {X.shape}")
print(f"   ‚Ä¢ Farms: {X.shape[0]}")
print(f"   ‚Ä¢ Features: {X.shape[1]}")
```

```python
      # ============================================================================
# CELL 4: FEATURE SCALING
# ============================================================================

print("\n" + "=" * 80)
print("‚öñÔ∏è FEATURE SCALING")
print("=" * 80)

print("Standardizing features for clustering...")

# Initialize and fit scaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert back to DataFrame for easier handling
X_scaled_df = pd.DataFrame(X_scaled, index=X.index, columns=X.columns)

print("‚úÖ Features scaled successfully!")

print(f"\nüìà Scaled feature statistics:")
scaled_stats = pd.DataFrame(X_scaled, columns=X.columns).describe().loc[['mean', 'std']].round(3)
print(scaled_stats) 
```

```python
# ============================================================================
# CELL 5: DETERMINE OPTIMAL NUMBER OF ZONES
# ============================================================================

print("\n" + "=" * 80)
print("üîç OPTIMAL ZONE DETERMINATION")
print("=" * 80)

print("Finding optimal number of rainfall zones...")

# Test different numbers of zones
silhouette_scores = []
zone_range = range(3, 8)  # Test 3 to 7 zones

print("\nTesting zone configurations:")
for n_zones in zone_range:
    kmeans = KMeans(n_clusters=n_zones, random_state=42, n_init=10)
    zone_labels = kmeans.fit_predict(X_scaled)
    
    if len(set(zone_labels)) > 1:
        score = silhouette_score(X_scaled, zone_labels)
        silhouette_scores.append(score)
        print(f"  ‚Ä¢ {n_zones} zones: silhouette score = {score:.3f}")
    else:
        silhouette_scores.append(0)
        print(f"  ‚Ä¢ {n_zones} zones: cannot compute (single cluster)")

# Find optimal number
optimal_zones = list(zone_range)[np.argmax(silhouette_scores)]
best_score = max(silhouette_scores)

print(f"\nüéØ OPTIMAL CONFIGURATION:")
print(f"   ‚Ä¢ Zones: {optimal_zones}")
print(f"   ‚Ä¢ Silhouette score: {best_score:.3f}")
print(f"   ‚Ä¢ Interpretation: {interpret_silhouette(best_score)}")

# Helper function
def interpret_silhouette(score):
    if score > 0.7:
        return "Strong structure"
    elif score > 0.5:
        return "Reasonable structure"
    elif score > 0.25:
        return "Weak structure"
    else:
        return "No substantial structure"

# Visualization
plt.figure(figsize=(10, 6))
bars = plt.bar(list(zone_range), silhouette_scores, color='steelblue', alpha=0.7)
plt.xlabel('Number of Zones', fontsize=12)
plt.ylabel('Silhouette Score', fontsize=12)
plt.title('Optimal Number of Rainfall Zones', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3, axis='y')

# Highlight optimal
bars[optimal_zones - 3].set_color('red')
bars[optimal_zones - 3].set_alpha(1.0)
plt.axhline(y=best_score, color='red', linestyle='--', alpha=0.5)
plt.text(optimal_zones, best_score + 0.01, f'Optimal: {optimal_zones} zones', 
         ha='center', fontweight='bold', color='red')

plt.tight_layout()
plt.show()
```

```python
# ============================================================================
# CELL 6: TRAIN ZONE CLASSIFICATION MODEL
# ============================================================================

print("\n" + "=" * 80)
print("ü§ñ TRAINING ZONE MODEL")
print("=" * 80)

print(f"Training final model with {optimal_zones} zones...")

# Train final model
zone_model = KMeans(n_clusters=optimal_zones, random_state=42, n_init=10)
zone_labels = zone_model.fit_predict(X_scaled)

# Add zones to features dataframe
features_df['zone_id'] = zone_labels
features_df['zone_name'] = [f'Zone_{z+1}' for z in zone_labels]

print("‚úÖ Model trained successfully!")

# Zone distribution
print(f"\nüìä ZONE DISTRIBUTION:")
zone_counts = features_df['zone_name'].value_counts().sort_index()
for zone_name, count in zone_counts.items():
    percentage = (count / len(features_df)) * 100
    print(f"  ‚Ä¢ {zone_name}: {count} farms ({percentage:.1f}%)")

# Zone centers (original scale)
zone_centers = pd.DataFrame(
    scaler.inverse_transform(zone_model.cluster_centers_),
    columns=available_features,
    index=[f'Zone_{i+1}' for i in range(optimal_zones)]
)

print(f"\nüìç ZONE CHARACTERISTICS (key metrics):")
print(zone_centers[['total_rainfall', 'rainy_days_percent', 'max_dry_spell', 'cv_rainfall']].round(2))
```

```python
# ============================================================================
# CELL 7: CREATE ZONE PROFILES
# ============================================================================

print("\n" + "=" * 80)
print("üìù ZONE PROFILES CREATION")
print("=" * 80)

print("Creating detailed profiles for each zone...")

def create_zone_profiles(features_df, optimal_zones):
    """Create comprehensive profiles for each rainfall zone"""
    zone_profiles = {}
    
    for zone_id in range(optimal_zones):
        zone_data = features_df[features_df['zone_id'] == zone_id]
        
        profile = {
            'zone_id': int(zone_id),
            'zone_name': f'Zone_{zone_id+1}',
            'n_farms': int(len(zone_data)),
            'percentage': float((len(zone_data) / len(features_df)) * 100)
        }
        
        # Calculate key statistics
        key_metrics = ['total_rainfall', 'mean_rainfall', 'rainy_days_percent', 
                      'max_dry_spell', 'cv_rainfall', 'seasonality_index']
        
        for metric in key_metrics:
            if metric in zone_data.columns:
                profile[f'avg_{metric}'] = float(zone_data[metric].mean())
                profile[f'std_{metric}'] = float(zone_data[metric].std())
        
        # Classify zone type
        profile['zone_type'] = classify_zone_type(profile)
        
        # Classify drought risk
        profile['drought_risk'] = classify_drought_risk(profile)
        
        # Classify rainfall regime
        profile['rainfall_regime'] = classify_rainfall_regime(profile)
        
        # Monthly pattern
        monthly_pattern = {}
        for month in range(1, 13):
            col_name = f'month_{month}_mean'
            if col_name in zone_data.columns:
                monthly_pattern[str(month)] = float(zone_data[col_name].mean())
        profile['monthly_pattern'] = monthly_pattern
        
        # Determine wettest and driest months
        if monthly_pattern:
            wettest_month = int(max(monthly_pattern, key=monthly_pattern.get))
            driest_month = int(min(monthly_pattern, key=monthly_pattern.get))
            profile['wettest_month'] = wettest_month
            profile['driest_month'] = driest_month
        
        # Insurance factors
        profile['premium_multiplier'] = calculate_premium_multiplier(profile['drought_risk'])
        profile['risk_score'] = calculate_risk_score(profile)
        
        zone_profiles[zone_id] = profile
    
    return zone_profiles

def classify_zone_type(profile):
    """Classify zone based on rainfall characteristics"""
    total_rain = profile.get('avg_total_rainfall', 0)
    rainy_days = profile.get('avg_rainy_days_percent', 0)
    
    if total_rain > 8000:
        if rainy_days > 60:
            return "High Rainfall, Frequent"
        else:
            return "High Rainfall, Intense"
    elif total_rain > 4000:
        if rainy_days > 50:
            return "Moderate Rainfall, Regular"
        else:
            return "Moderate Rainfall, Seasonal"
    elif total_rain > 1500:
        return "Low Rainfall Zone"
    else:
        return "Arid/Semi-Arid Zone"

def classify_drought_risk(profile):
    """Classify drought risk level"""
    risk_score = 0
    
    # Rainfall amount
    total_rain = profile.get('avg_total_rainfall', 0)
    if total_rain < 1000:
        risk_score += 3
    elif total_rain < 3000:
        risk_score += 2
    elif total_rain < 5000:
        risk_score += 1
    
    # Variability
    cv_rain = profile.get('avg_cv_rainfall', 0)
    if cv_rain > 1.0:
        risk_score += 2
    elif cv_rain > 0.7:
        risk_score += 1
    
    # Dry spells
    max_dry = profile.get('avg_max_dry_spell', 0)
    if max_dry > 200:
        risk_score += 2
    elif max_dry > 100:
        risk_score += 1
    
    # Rainy days
    rainy_pct = profile.get('avg_rainy_days_percent', 0)
    if rainy_pct < 15:
        risk_score += 2
    elif rainy_pct < 30:
        risk_score += 1
    
    # Classify
    if risk_score >= 7:
        return "Very High"
    elif risk_score >= 5:
        return "High"
    elif risk_score >= 3:
        return "Moderate"
    else:
        return "Low"

def classify_rainfall_regime(profile):
    """Classify rainfall regime"""
    seasonality = profile.get('avg_seasonality_index', 0)
    
    if seasonality > 1.2:
        return "Highly Seasonal"
    elif seasonality > 0.8:
        return "Seasonal"
    elif seasonality > 0.4:
        return "Moderately Seasonal"
    else:
        return "Uniform"

def calculate_premium_multiplier(drought_risk):
    """Calculate insurance premium multiplier"""
    multipliers = {
        "Very High": 2.5,
        "High": 1.8,
        "Moderate": 1.3,
        "Low": 1.0
    }
    return multipliers.get(drought_risk, 1.3)

def calculate_risk_score(profile):
    """Calculate overall risk score (0-10)"""
    base_score = 0
    
    drought_risk = profile.get('drought_risk', 'Moderate')
    if drought_risk == "Very High":
        base_score += 8
    elif drought_risk == "High":
        base_score += 6
    elif drought_risk == "Moderate":
        base_score += 4
    else:
        base_score += 2
    
    # Adjust for variability
    cv = profile.get('avg_cv_rainfall', 0)
    if cv > 0.8:
        base_score += 2
    
    return min(10, base_score)

# Create zone profiles
zone_profiles = create_zone_profiles(features_df, optimal_zones)

print("‚úÖ Zone profiles created successfully!")

print(f"\nüìã ZONE PROFILES SUMMARY:")
for zone_id, profile in zone_profiles.items():
    print(f"\n  {profile['zone_name']} ({profile['zone_type']}):")
    print(f"    ‚Ä¢ Farms: {profile['n_farms']} ({profile['percentage']:.1f}%)")
    print(f"    ‚Ä¢ Avg Rainfall: {profile.get('avg_total_rainfall', 0):.0f} mm")
    print(f"    ‚Ä¢ Rainy Days: {profile.get('avg_rainy_days_percent', 0):.1f}%")
    print(f"    ‚Ä¢ Drought Risk: {profile['drought_risk']}")
    print(f"    ‚Ä¢ Premium Multiplier: {profile['premium_multiplier']}x")
    print(f"    ‚Ä¢ Risk Score: {profile['risk_score']}/10")
```

```python
# ============================================================================
# CELL 10: SAVE MODEL AND FILES
# ============================================================================

print("\n" + "=" * 80)
print("üíæ SAVING MODEL AND FILES")
print("=" * 80)

import datetime
import os

print("Saving model artifacts for production...")

# Create output directory
output_dir = 'rainfall_zone_model'
os.makedirs(output_dir, exist_ok=True)

# Custom JSON encoder for numpy types
class NumpyEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        elif isinstance(obj, np.floating):
            return float(obj)
        elif isinstance(obj, np.ndarray):
            return obj.tolist()
        elif isinstance(obj, np.bool_):
            return bool(obj)
        return super().default(obj)

# Prepare model data
model_data = {
    'zone_model': zone_model,
    'scaler': scaler,
    'feature_names': available_features,
    'zone_profiles': zone_profiles,
    'model_info': {
        'n_zones': optimal_zones,
        'n_farms': len(features_df),
        'training_date': datetime.datetime.now().isoformat(),
        'features_used': available_features,
        'model_type': 'KMeans Clustering',
        'silhouette_score': float(max(silhouette_scores)),
        'zone_distribution': zone_counts.to_dict()
    }
}

# 1. Save main model as pickle
model_path = os.path.join(output_dir, 'rainfall_zone_model.pkl')
with open(model_path, 'wb') as f:
    pickle.dump(model_data, f)
print(f"‚úÖ 1. Main model saved: {model_path}")

# 2. Save zone profiles as JSON
profiles_path = os.path.join(output_dir, 'zone_profiles.json')
with open(profiles_path, 'w') as f:
    json.dump(zone_profiles, f, indent=2, cls=NumpyEncoder)
print(f"‚úÖ 2. Zone profiles saved: {profiles_path}")

# 3. Save farm classifications
farms_path = os.path.join(output_dir, 'farm_classifications.csv')
farm_classifications = features_df[['zone_id', 'zone_name']].copy()
farm_classifications.index.name = 'farm_id'
farm_classifications.to_csv(farms_path)
print(f"‚úÖ 3. Farm classifications saved: {farms_path}")

# 4. Save feature summary
summary_path = os.path.join(output_dir, 'feature_summary.csv')
feature_summary = features_df.describe().round(3)
feature_summary.to_csv(summary_path)
print(f"‚úÖ 4. Feature summary saved: {summary_path}")

# 5. Save prediction function
prediction_script = '''
import pickle
import numpy as np
import json

class RainfallZonePredictor:
    """Production-ready rainfall zone predictor"""
    
    def __init__(self, model_path='rainfall_zone_model.pkl'):
        """Load trained model"""
        with open(model_path, 'rb') as f:
            self.model_data = pickle.load(f)
        
        self.zone_model = self.model_data['zone_model']
        self.scaler = self.model_data['scaler']
        self.feature_names = self.model_data['feature_names']
        self.zone_profiles = self.model_data['zone_profiles']
        
    def predict(self, monthly_rainfall, farm_id="new_farm"):
        """Predict zone for new farm"""
        # Calculate features
        features = self._calculate_features(monthly_rainfall)
        
        # Prepare and scale features
        feature_vector = self._prepare_features(features)
        feature_vector_scaled = self.scaler.transform([feature_vector])
        
        # Predict
        zone_id = int(self.zone_model.predict(feature_vector_scaled)[0])
        
        # Get results
        return self._format_result(zone_id, features, farm_id)
    
    def _calculate_features(self, rainfall):
        """Calculate features from monthly rainfall"""
        features = {}
        rainfall = np.array(rainfall)
        
        # Basic stats
        features['total_rainfall'] = float(np.sum(rainfall))
        features['mean_rainfall'] = float(np.mean(rainfall))
        features['std_rainfall'] = float(np.std(rainfall))
        
        if features['mean_rainfall'] > 0:
            features['cv_rainfall'] = float(features['std_rainfall'] / features['mean_rainfall'])
        else:
            features['cv_rainfall'] = 0.0
        
        # Rainy days
        rainy_days = rainfall[rainfall > 0]
        features['rainy_days_percent'] = float((len(rainy_days) / 12) * 100)
        
        # Dry spells
        dry_spells = []
        current = 0
        for rain in rainfall:
            if rain == 0:
                current += 1
            else:
                if current > 0:
                    dry_spells.append(current)
                current = 0
        if current > 0:
            dry_spells.append(current)
        
        features['max_dry_spell'] = int(max(dry_spells)) if dry_spells else 0
        
        # Monthly means
        for i, rain in enumerate(rainfall, 1):
            features[f'month_{i}_mean'] = float(rain)
        
        return features
    
    def _prepare_features(self, features):
        """Prepare feature vector in correct order"""
        vector = []
        for name in self.feature_names:
            vector.append(features.get(name, 0.0))
        return vector
    
    def _format_result(self, zone_id, features, farm_id):
        """Format prediction result"""
        profile = self.zone_profiles.get(zone_id, {})
        
        return {
            'farm_id': farm_id,
            'zone_id': zone_id,
            'zone_name': profile.get('zone_name', f'Zone_{zone_id+1}'),
            'zone_type': profile.get('zone_type', 'Unknown'),
            'drought_risk': profile.get('drought_risk', 'Unknown'),
            'total_rainfall': features['total_rainfall'],
            'rainy_days_percent': features['rainy_days_percent'],
            'premium_multiplier': profile.get('premium_multiplier', 1.3),
            'risk_score': profile.get('risk_score', 5)
        }

# Example usage:
# predictor = RainfallZonePredictor('rainfall_zone_model.pkl')
# result = predictor.predict([50, 40, 60, 80, 100, 20, 10, 5, 15, 40, 120, 90])
'''

predictor_path = os.path.join(output_dir, 'predictor.py')
with open(predictor_path, 'w') as f:
    f.write(prediction_script)
print(f"‚úÖ 5. Predictor script saved: {predictor_path}")

# 6. Save README file
readme_content = f"""
RAINFALL ZONE CLASSIFICATION MODEL
====================================

Model Information:
------------------
‚Ä¢ Training Date: {datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')}
‚Ä¢ Number of Zones: {optimal_zones}
‚Ä¢ Number of Farms: {len(features_df)}
‚Ä¢ Model Type: KMeans Clustering
‚Ä¢ Silhouette Score: {max(silhouette_scores):.3f}

Files:
------
1. rainfall_zone_model.pkl - Main model file (pickle format)
2. zone_profiles.json - Zone descriptions and characteristics
3. farm_classifications.csv - All farm classifications
4. feature_summary.csv - Statistical summary of features
5. predictor.py - Ready-to-use prediction class

Zone Distribution:
-----------------
{zone_counts.to_string()}

How to Use:
-----------
1. Load the model: predictor = RainfallZonePredictor('rainfall_zone_model.pkl')
2. Predict zone: result = predictor.predict([12 monthly values])
3. Results include zone, drought risk, and premium multiplier

For Insurance Companies:
-----------------------
‚Ä¢ Use premium_multiplier for pricing
‚Ä¢ Use drought_risk for risk assessment
‚Ä¢ Use zone_type for product segmentation
"""

readme_path = os.path.join(output_dir, 'README.md')
with open(readme_path, 'w') as f:
    f.write(readme_content)
print(f"‚úÖ 6. README file saved: {readme_path}")

print("\n" + "=" * 80)
print("üéâ MODEL TRAINING COMPLETE!")
print("=" * 80)

print(f"\nüìä FINAL SUMMARY:")
print(f"   ‚Ä¢ Farms classified: {len(features_df)}")
print(f"   ‚Ä¢ Rainfall zones created: {optimal_zones}")
print(f"   ‚Ä¢ Model quality: {max(silhouette_scores):.3f} silhouette score")
print(f"   ‚Ä¢ Files saved to: {output_dir}/")
print(f"   ‚Ä¢ Ready for API integration")

print("\nüìã NEXT STEPS:")
print("   1. Deploy predictor.py with your backend API")
print("   2. Use zone_profiles.json in frontend dashboards")
print("   3. Integrate with farmer registration system")
print("   4. Update model periodically with new data")
```
