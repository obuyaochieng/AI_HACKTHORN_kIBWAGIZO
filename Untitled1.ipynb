# Regional Classification On Rainfall Basis

comprehensive classification system that groups areas based on rainfall patterns and drought risk across different time scales. This will help you understand regional rainfall climatology and vulnerability

# SETUP AND DATA PREPARATION FOR CLASSIFICATION

This cell sets up the environment and loads necessary clustering libraries for pattern classification.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, calinski_harabasz_score
import warnings
warnings.filterwarnings('ignore')

print("=" * 80)
print("RAINFALL PATTERN CLASSIFICATION AND DROUGHT RISK ASSESSMENT")
print("=" * 80)

# If df is not already loaded:
df = pd.read_csv('output.csv')

print(f"Dataset shape: {df.shape}")
print(f"Unique locations: {df['id_number'].nunique()}")
```

# CREATE AGGREGATED FEATURES FOR CLASSIFICATION
 Creates comprehensive features for each location including drought metrics, rainfall statistics, and temporal trends.

```python
# ============================================================================
# CELL 2: CREATE AGGREGATED FEATURES FOR CLASSIFICATION (FIXED)
# ============================================================================

print("\n2. CREATING AGGREGATED FEATURES FOR CLASSIFICATION...")

# Function to calculate drought indices
def calculate_drought_metrics(group):
    """Calculate various drought-related metrics for a location"""
    metrics = {}
    
    # Basic statistics
    metrics['total_rainfall'] = group['rainfall_mm'].sum()
    metrics['mean_rainfall'] = group['rainfall_mm'].mean()
    metrics['median_rainfall'] = group['rainfall_mm'].median()
    metrics['std_rainfall'] = group['rainfall_mm'].std()
    metrics['cv_rainfall'] = metrics['std_rainfall'] / metrics['mean_rainfall'] if metrics['mean_rainfall'] > 0 else 0
    
    # Rainy days metrics
    rainy_days = group[group['rainfall_mm'] > 0]
    metrics['rainy_days_count'] = len(rainy_days)
    metrics['rainy_days_percent'] = (len(rainy_days) / len(group)) * 100 if len(group) > 0 else 0
    metrics['mean_rainy_day_intensity'] = rainy_days['rainfall_mm'].mean() if len(rainy_days) > 0 else 0
    
    # Dry spell metrics
    dry_spells = []
    current_spell = 0
    for val in group['rainfall_mm'].values:
        if val == 0:
            current_spell += 1
        else:
            if current_spell > 0:
                dry_spells.append(current_spell)
            current_spell = 0
    if current_spell > 0:
        dry_spells.append(current_spell)
    
    metrics['max_dry_spell'] = max(dry_spells) if dry_spells else 0
    metrics['avg_dry_spell'] = np.mean(dry_spells) if dry_spells else 0
    metrics['dry_spell_frequency'] = len(dry_spells) / len(group) if len(group) > 0 else 0
    
    # Extreme events
    if len(group) > 0:
        metrics['p95_events'] = len(group[group['rainfall_mm'] > group['rainfall_mm'].quantile(0.95)])
        metrics['p99_events'] = len(group[group['rainfall_mm'] > group['rainfall_mm'].quantile(0.99)])
    else:
        metrics['p95_events'] = 0
        metrics['p99_events'] = 0
    
    # Seasonality metrics (using month data)
    if 'month' in group.columns and len(group) > 0:
        monthly_means = group.groupby('month')['rainfall_mm'].mean()
        if len(monthly_means) > 0 and monthly_means.mean() > 0:
            metrics['seasonality_index'] = monthly_means.std() / monthly_means.mean()
            metrics['wet_month_ratio'] = len(monthly_means[monthly_means > monthly_means.mean()]) / len(monthly_means)
        else:
            metrics['seasonality_index'] = 0
            metrics['wet_month_ratio'] = 0
    else:
        metrics['seasonality_index'] = 0
        metrics['wet_month_ratio'] = 0
    
    return pd.Series(metrics)

# Create aggregated features for each location
print("Creating location-level features...")
location_features = df.groupby('id_number').apply(calculate_drought_metrics)

# Add yearly trend metrics
print("Calculating temporal trends...")
yearly_trends = []
for location_id in df['id_number'].unique():
    location_data = df[df['id_number'] == location_id]
    yearly_rain = location_data.groupby('year')['rainfall_mm'].sum()
    
    if len(yearly_rain) > 1:
        # Linear trend
        x = np.array(yearly_rain.index)
        y = yearly_rain.values
        z = np.polyfit(x, y, 1)
        trend_slope = z[0]
        trend_pct = (z[0] * len(x)) / np.mean(y) * 100 if np.mean(y) > 0 else 0
    else:
        trend_slope = 0
        trend_pct = 0
    
    yearly_trends.append({
        'id_number': location_id,
        'yearly_trend_slope': trend_slope,
        'yearly_trend_pct': trend_pct,
        'yearly_rainfall_cv': np.std(yearly_rain) / np.mean(yearly_rain) if np.mean(yearly_rain) > 0 else 0
    })

yearly_trends_df = pd.DataFrame(yearly_trends).set_index('id_number')
location_features = location_features.join(yearly_trends_df)

print(f"Created {len(location_features.columns)} features for {len(location_features)} locations")
print("\nSample features:")
print(location_features.head())
print(f"\nFeature columns: {list(location_features.columns)}")
```

# CREATE TIME-SPECIFIC FEATURES (MONTHLY, QUARTERLY, SEASONAL)

```python
# ============================================================================
# CELL 3: CREATE TIME-SPECIFIC FEATURES (MONTHLY, QUARTERLY, SEASONAL) - FIXED
# ============================================================================

print("\n3. CREATING TIME-SPECIFIC FEATURES...")

# First, let's check what columns we have in df
print(f"Available columns in df: {list(df.columns)}")

# Ensure we have the necessary temporal features
if 'month' not in df.columns:
    print("Warning: 'month' column not found in df")
    # If you have a date column, you can extract month from it:
    # df['month'] = pd.to_datetime(df['date_column']).dt.month

# Monthly features
print("\nExtracting monthly patterns...")
monthly_features_list = []

for location_id in df['id_number'].unique():
    location_data = df[df['id_number'] == location_id]
    
    # Calculate monthly statistics
    monthly_stats = location_data.groupby('month')['rainfall_mm'].agg(['mean', 'std', 'count'])
    
    # Create a flat structure for the features
    monthly_dict = {'id_number': location_id}
    
    for month in range(1, 13):
        if month in monthly_stats.index:
            month_data = monthly_stats.loc[month]
            monthly_dict[f'month_{month}_mean'] = month_data['mean']
            monthly_dict[f'month_{month}_std'] = month_data['std']
            monthly_dict[f'month_{month}_count'] = month_data['count']
        else:
            monthly_dict[f'month_{month}_mean'] = 0
            monthly_dict[f'month_{month}_std'] = 0
            monthly_dict[f'month_{month}_count'] = 0
    
    monthly_features_list.append(monthly_dict)

monthly_features = pd.DataFrame(monthly_features_list).set_index('id_number')

print(f"Monthly features created: {monthly_features.shape}")

# Quarterly features
print("\nExtracting quarterly patterns...")
# First, ensure we have quarter column
if 'quarter' not in df.columns:
    df['quarter'] = df['month'].apply(lambda x: f"Q{(x-1)//3 + 1}")

quarterly_features_list = []

for location_id in df['id_number'].unique():
    location_data = df[df['id_number'] == location_id]
    
    # Calculate quarterly statistics
    quarterly_stats = location_data.groupby('quarter')['rainfall_mm'].agg(['mean', 'sum', 'std'])
    
    # Create a flat structure
    quarterly_dict = {'id_number': location_id}
    
    # Define quarters in order
    quarters = ['Q1', 'Q2', 'Q3', 'Q4']
    
    for quarter in quarters:
        if quarter in quarterly_stats.index:
            quarter_data = quarterly_stats.loc[quarter]
            quarterly_dict[f'quarter_{quarter}_mean'] = quarter_data['mean']
            quarterly_dict[f'quarter_{quarter}_sum'] = quarter_data['sum']
            quarterly_dict[f'quarter_{quarter}_std'] = quarter_data['std']
        else:
            quarterly_dict[f'quarter_{quarter}_mean'] = 0
            quarterly_dict[f'quarter_{quarter}_sum'] = 0
            quarterly_dict[f'quarter_{quarter}_std'] = 0
    
    quarterly_features_list.append(quarterly_dict)

quarterly_features = pd.DataFrame(quarterly_features_list).set_index('id_number')

print(f"Quarterly features created: {quarterly_features.shape}")

# Seasonal features
print("\nExtracting seasonal patterns...")
# First, ensure we have season column
if 'season' not in df.columns:
    # Define seasons
    def get_season(month):
        if month in [12, 1, 2]:
            return 'DJF'
        elif month in [3, 4, 5]:
            return 'MAM'
        elif month in [6, 7, 8]:
            return 'JJA'
        else:  # 9, 10, 11
            return 'SON'
    
    df['season'] = df['month'].apply(get_season)

seasonal_features_list = []

for location_id in df['id_number'].unique():
    location_data = df[df['id_number'] == location_id]
    
    # Calculate seasonal statistics
    seasonal_stats = location_data.groupby('season')['rainfall_mm'].agg(['mean', 'sum', 'std'])
    
    # Create a flat structure
    seasonal_dict = {'id_number': location_id}
    
    # Define seasons in order
    seasons = ['DJF', 'MAM', 'JJA', 'SON']
    
    for season in seasons:
        if season in seasonal_stats.index:
            season_data = seasonal_stats.loc[season]
            seasonal_dict[f'season_{season}_mean'] = season_data['mean']
            seasonal_dict[f'season_{season}_sum'] = season_data['sum']
            seasonal_dict[f'season_{season}_std'] = season_data['std']
        else:
            seasonal_dict[f'season_{season}_mean'] = 0
            seasonal_dict[f'season_{season}_sum'] = 0
            seasonal_dict[f'season_{season}_std'] = 0
    
    seasonal_features_list.append(seasonal_dict)

seasonal_features = pd.DataFrame(seasonal_features_list).set_index('id_number')

print(f"Seasonal features created: {seasonal_features.shape}")

# Combine all features
print("\nCombining all features...")
all_features = location_features.copy()

# Join monthly features
all_features = all_features.join(monthly_features, how='left')

# Join quarterly features
all_features = all_features.join(quarterly_features, how='left')

# Join seasonal features
all_features = all_features.join(seasonal_features, how='left')

# Fill NaN values
all_features = all_features.fillna(0)

print(f"\nFinal feature matrix shape: {all_features.shape}")
print(f"Total features created: {len(all_features.columns)}")
print(f"\nFirst few feature names: {list(all_features.columns)[:20]}...")
```

 # FEATURE SELECTION AND SCALING

```python
# ============================================================================
# CELL 4: FEATURE SELECTION AND SCALING
# ============================================================================

print("\n4. FEATURE SELECTION AND PREPROCESSING...")

# Let's first check what features we actually have
print(f"Available features: {len(all_features.columns)}")
print("\nSample features:")
print(list(all_features.columns)[:30])  # Show first 30 features

# Select key features for clustering
# Based on your data, I'll select the most relevant features
selected_features = []

# Check which features are available and select them
possible_features = [
    'total_rainfall', 'mean_rainfall', 'cv_rainfall',
    'rainy_days_percent', 'mean_rainy_day_intensity',
    'max_dry_spell', 'avg_dry_spell', 'dry_spell_frequency',
    'p95_events', 'p99_events', 'seasonality_index',
    'yearly_trend_slope', 'yearly_trend_pct', 'yearly_rainfall_cv'
]

for feature in possible_features:
    if feature in all_features.columns:
        selected_features.append(feature)
        print(f"✓ Selected: {feature}")
    else:
        print(f"✗ Not available: {feature}")

# Add key monthly features (focus on critical months based on your analysis)
# Your analysis showed Nov and Dec are wettest, Jun is driest
critical_months = {
    'Nov': 11,  # Wettest month
    'Dec': 12,  # Very wet
    'Jun': 6,   # Driest month
    'Mar': 3,   # Also wet
    'Apr': 4    # Also wet
}

for month_name, month_num in critical_months.items():
    mean_col = f'month_{month_num}_mean'
    sum_col = f'month_{month_num}_sum'  # We might not have this, but let's check
    if mean_col in all_features.columns:
        selected_features.append(mean_col)
        print(f"✓ Selected monthly feature: {mean_col} ({month_name})")

# Add key quarterly and seasonal features
# Your analysis showed Q4 is wettest, Q3 is driest
quarterly_to_check = ['Q4_sum', 'Q3_sum']
for q_feature in quarterly_to_check:
    col = f'quarter_{q_feature}'
    if col in all_features.columns:
        selected_features.append(col)
        print(f"✓ Selected quarterly feature: {col}")

# Seasonal features based on your analysis
seasonal_to_check = ['DJF_sum', 'JJA_sum']
for s_feature in seasonal_to_check:
    col = f'season_{s_feature}'
    if col in all_features.columns:
        selected_features.append(col)
        print(f"✓ Selected seasonal feature: {col}")

print(f"\nTotal selected features: {len(selected_features)}")
print(f"\nSelected features: {selected_features}")

# Filter to available features and create feature matrix
available_features = [f for f in selected_features if f in all_features.columns]
X = all_features[available_features].copy()

print(f"\nFinal feature matrix shape: {X.shape}")
print(f"Number of locations: {len(X)}")

# Check for any infinite or NaN values
print(f"\nChecking data quality:")
print(f"NaN values: {X.isna().sum().sum()}")
print(f"Infinite values: {np.isinf(X.values).sum()}")

# Replace any NaN or infinite values
X = X.replace([np.inf, -np.inf], 0)
X = X.fillna(0)

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, index=X.index, columns=X.columns)

print("\nFeature statistics after scaling:")
print(pd.DataFrame(X_scaled, columns=X.columns).describe().loc[['mean', 'std']].round(3))
```

# DETERMINE OPTIMAL NUMBER OF CLUSTERS

```python
# ============================================================================
# CELL 5: DETERMINE OPTIMAL NUMBER OF CLUSTERS
# ============================================================================

print("\n5. DETERMINING OPTIMAL NUMBER OF CLUSTERS...")

# Evaluate different numbers of clusters
silhouette_scores = []
calinski_scores = []
cluster_range = range(2, min(11, len(X) // 2))  # Don't exceed half the samples

for n_clusters in cluster_range:
    try:
        kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
        cluster_labels = kmeans.fit_predict(X_scaled)
        
        if len(set(cluster_labels)) > 1:
            silhouette_scores.append(silhouette_score(X_scaled, cluster_labels))
            calinski_scores.append(calinski_harabasz_score(X_scaled, cluster_labels))
        else:
            silhouette_scores.append(0)
            calinski_scores.append(0)
    except Exception as e:
        print(f"Warning: Could not compute for {n_clusters} clusters: {e}")
        silhouette_scores.append(0)
        calinski_scores.append(0)

# Plot evaluation metrics
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(list(cluster_range), silhouette_scores, marker='o', linewidth=2, color='blue')
axes[0].set_xlabel('Number of Clusters')
axes[0].set_ylabel('Silhouette Score')
axes[0].set_title('Silhouette Score vs Number of Clusters')
axes[0].grid(True, alpha=0.3)

axes[1].plot(list(cluster_range), calinski_scores, marker='s', linewidth=2, color='green')
axes[1].set_xlabel('Number of Clusters')
axes[1].set_ylabel('Calinski-Harabasz Score')
axes[1].set_title('Calinski-Harabasz Score vs Number of Clusters')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Choose optimal number of clusters
if len(silhouette_scores) > 0:
    optimal_clusters = list(cluster_range)[np.argmax(silhouette_scores)]
    print(f"\nOptimal number of clusters based on silhouette score: {optimal_clusters}")
    print(f"Best silhouette score: {max(silhouette_scores):.3f}")
else:
    print("Could not determine optimal clusters. Using 4 as default.")
    optimal_clusters = 4
```

# APPLY K-MEANS CLUSTERING

```python
# ============================================================================
# CELL 6: APPLY K-MEANS CLUSTERING
# ============================================================================

print("\n6. APPLYING K-MEANS CLUSTERING...")

# Apply K-means with optimal clusters
try:
    kmeans = KMeans(n_clusters=optimal_clusters, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X_scaled)
    
    # Add cluster labels to features
    all_features['cluster'] = cluster_labels
    all_features['cluster_name'] = ['Cluster_' + str(i) for i in cluster_labels]
    
    # Calculate cluster centers
    cluster_centers = pd.DataFrame(
        scaler.inverse_transform(kmeans.cluster_centers_),
        columns=X.columns,
        index=[f'Cluster_{i}' for i in range(optimal_clusters)]
    )
    
    print(f"\nClustering complete! Assigned {optimal_clusters} clusters.")
    print(f"\nDistribution of locations across clusters:")
    cluster_counts = all_features['cluster'].value_counts().sort_index()
    for cluster_id, count in cluster_counts.items():
        percentage = (count / len(all_features)) * 100
        print(f"  Cluster {cluster_id}: {count} locations ({percentage:.1f}%)")
    
    print("\nCluster centers (original scale):")
    print(cluster_centers.round(2))
    
except Exception as e:
    print(f"Error in clustering: {e}")
    # Fallback: assign all to same cluster
    all_features['cluster'] = 0
    all_features['cluster_name'] = 'Cluster_0'
    print("Assigned all locations to single cluster as fallback.")
```

# DROUGHT RISK CLASSIFICATION

```python
# ============================================================================
# CELL 8: DROUGHT RISK CLASSIFICATION
# ============================================================================

print("\n8. CLASSIFYING DROUGHT RISK LEVELS...")

# Define drought risk based on multiple factors
def classify_drought_risk(row):
    """Classify drought risk based on multiple indicators"""
    risk_score = 0
    
    # Factor 1: Low rainfall
    total_rain_q25 = all_features['total_rainfall'].quantile(0.25)
    total_rain_q50 = all_features['total_rainfall'].quantile(0.5)
    
    if row['total_rainfall'] < total_rain_q25:
        risk_score += 2
    elif row['total_rainfall'] < total_rain_q50:
        risk_score += 1
    
    # Factor 2: High variability
    if 'cv_rainfall' in row:
        cv_q75 = all_features['cv_rainfall'].quantile(0.75)
        cv_q50 = all_features['cv_rainfall'].quantile(0.5)
        
        if row['cv_rainfall'] > cv_q75:
            risk_score += 2
        elif row['cv_rainfall'] > cv_q50:
            risk_score += 1
    
    # Factor 3: Long dry spells
    if 'max_dry_spell' in row:
        dry_spell_q75 = all_features['max_dry_spell'].quantile(0.75)
        dry_spell_q50 = all_features['max_dry_spell'].quantile(0.5)
        
        if row['max_dry_spell'] > dry_spell_q75:
            risk_score += 2
        elif row['max_dry_spell'] > dry_spell_q50:
            risk_score += 1
    
    # Factor 4: Low rainy days
    if 'rainy_days_percent' in row:
        rainy_q25 = all_features['rainy_days_percent'].quantile(0.25)
        rainy_q50 = all_features['rainy_days_percent'].quantile(0.5)
        
        if row['rainy_days_percent'] < rainy_q25:
            risk_score += 2
        elif row['rainy_days_percent'] < rainy_q50:
            risk_score += 1
    
    # Factor 5: Negative trend
    if 'yearly_trend_slope' in row and row['yearly_trend_slope'] < 0:
        risk_score += 1
    
    # Classify based on total score
    if risk_score >= 7:
        return 'Very High'
    elif risk_score >= 5:
        return 'High'
    elif risk_score >= 3:
        return 'Moderate'
    else:
        return 'Low'

# Apply drought risk classification
all_features['drought_risk'] = all_features.apply(classify_drought_risk, axis=1)

print("\nDrought Risk Classification Results:")
risk_counts = all_features['drought_risk'].value_counts()
for risk_level, count in risk_counts.items():
    percentage = (count / len(all_features)) * 100
    print(f"  {risk_level}: {count} locations ({percentage:.1f}%)")

# Cross-tabulate clusters with drought risk
print("\nCluster vs Drought Risk Cross-tabulation:")
cross_tab = pd.crosstab(all_features['cluster'], all_features['drought_risk'])
print(cross_tab)

# Visualize the cross-tabulation
fig, ax = plt.subplots(figsize=(10, 6))
cross_tab.plot(kind='bar', stacked=True, ax=ax, colormap='RdYlGn_r')
ax.set_xlabel('Cluster')
ax.set_ylabel('Number of Locations')
ax.set_title('Drought Risk Distribution by Cluster')
ax.legend(title='Drought Risk', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# CREATE RAINFALL REGIME CLASSIFICATION

```python
# ============================================================================
# CELL 9: CREATE RAINFALL REGIME CLASSIFICATION
# ============================================================================

print("\n9. CREATING RAINFALL REGIME CLASSIFICATION...")

def classify_rainfall_regime(row):
    """Classify rainfall regime based on cluster characteristics"""
    cluster_id = row['cluster']
    
    # Get cluster averages
    cluster_data = all_features[all_features['cluster'] == cluster_id]
    
    # Calculate thresholds based on entire dataset
    total_rain_q75 = all_features['total_rainfall'].quantile(0.75)
    total_rain_q50 = all_features['total_rainfall'].quantile(0.5)
    total_rain_q25 = all_features['total_rainfall'].quantile(0.25)
    
    rainy_days_q75 = all_features['rainy_days_percent'].quantile(0.75)
    rainy_days_q25 = all_features['rainy_days_percent'].quantile(0.25)
    
    seasonality_q75 = all_features['seasonality_index'].quantile(0.75) if 'seasonality_index' in all_features.columns else 1
    
    cluster_avg_total = cluster_data['total_rainfall'].mean()
    cluster_avg_rainy_days = cluster_data['rainy_days_percent'].mean()
    
    # Determine which months are wettest for this cluster
    wet_months = []
    for month in range(1, 13):
        col_name = f'month_{month}_mean'
        if col_name in cluster_data.columns:
            month_avg = cluster_data[col_name].mean()
            if month_avg > cluster_data[f'month_{month}_mean'].mean() * 1.2:  # 20% above cluster average
                wet_months.append(month)
    
    # Classify based on characteristics
    if cluster_avg_total > total_rain_q75:
        if len(wet_months) >= 3:
            if any(m in [11, 12, 1] for m in wet_months):  # Nov-Dec-Jan
                regime = 'High Rainfall, Winter Wet'
            elif any(m in [6, 7, 8] for m in wet_months):  # Jun-Jul-Aug
                regime = 'High Rainfall, Summer Wet'
            else:
                regime = 'High Rainfall, Seasonal'
        else:
            regime = 'High Rainfall, Less Seasonal'
    
    elif cluster_avg_total > total_rain_q50:
        if cluster_avg_rainy_days > rainy_days_q75:
            regime = 'Moderate Rainfall, Frequent Rain'
        else:
            regime = 'Moderate Rainfall, Infrequent Rain'
    
    elif cluster_avg_total > total_rain_q25:
        if cluster_avg_rainy_days < rainy_days_q25:
            regime = 'Low Rainfall, Arid'
        else:
            regime = 'Low Rainfall, Seasonal'
    
    else:
        regime = 'Very Low Rainfall, Hyper-Arid'
    
    return regime

# Apply regime classification
all_features['rainfall_regime'] = all_features.apply(classify_rainfall_regime, axis=1)

print("\nRainfall Regime Classification:")
regime_counts = all_features['rainfall_regime'].value_counts()
for regime, count in regime_counts.items():
    percentage = (count / len(all_features)) * 100
    print(f"  {regime}: {count} locations ({percentage:.1f}%)")

# Create comprehensive classification
all_features['comprehensive_class'] = all_features.apply(
    lambda x: f"{x['rainfall_regime']} - {x['drought_risk']} Risk", axis=1)

print("\nTop 10 Comprehensive Classifications:")
print(all_features['comprehensive_class'].value_counts().head(10))
```

# CREATE MONTHLY, QUARTERLY, AND SEASONAL CLASSIFICATIONS

```python
# ============================================================================
# CELL 11: CREATE MONTHLY, QUARTERLY, AND SEASONAL CLASSIFICATIONS
# ============================================================================

print("\n11. CREATING TIME-SPECIFIC CLASSIFICATIONS...")

# Monthly classification
print("\nCreating monthly rainfall pattern classification...")
monthly_classification = {}

for location_id in all_features.index:
    location_data = all_features.loc[location_id]
    monthly_pattern = []
    
    for month in range(1, 13):
        col_name = f'month_{month}_mean'
        if col_name in all_features.columns:
            month_rain = location_data[col_name]
            # Use overall mean for this month across all locations
            overall_mean = all_features[col_name].mean()
            
            if month_rain > overall_mean * 1.5:
                monthly_pattern.append('Wet')
            elif month_rain < overall_mean * 0.5:
                monthly_pattern.append('Dry')
            else:
                monthly_pattern.append('Normal')
        else:
            monthly_pattern.append('Unknown')
    
    # Determine monthly pattern type
    wet_months = monthly_pattern.count('Wet')
    dry_months = monthly_pattern.count('Dry')
    
    if wet_months >= 6:
        monthly_type = 'Mostly Wet Year'
    elif dry_months >= 6:
        monthly_type = 'Mostly Dry Year'
    elif any(pattern == 'Wet' for pattern in monthly_pattern[10:] + monthly_pattern[:2]):  # Nov-Dec or Jan-Feb
        monthly_type = 'Wet Winter'
    elif any(pattern == 'Wet' for pattern in monthly_pattern[5:8]):  # Jun-Aug
        monthly_type = 'Wet Summer'
    elif any(pattern == 'Wet' for pattern in monthly_pattern[2:5]):  # Mar-May
        monthly_type = 'Wet Spring'
    else:
        monthly_type = 'Variable/Erratic'
    
    monthly_classification[location_id] = {
        'monthly_pattern': monthly_pattern,
        'monthly_type': monthly_type,
        'wet_months_count': wet_months,
        'dry_months_count': dry_months
    }

# Quarterly classification
print("\nCreating quarterly rainfall pattern classification...")
quarterly_classification = {}

for location_id in all_features.index:
    location_data = all_features.loc[location_id]
    quarterly_sums = {}
    
    for quarter in ['Q1', 'Q2', 'Q3', 'Q4']:
        sum_col = f'quarter_{quarter}_sum'
        if sum_col in all_features.columns:
            quarterly_sums[quarter] = location_data[sum_col]
        else:
            quarterly_sums[quarter] = 0
    
    # Find wettest and driest quarters
    if quarterly_sums:
        wettest_quarter = max(quarterly_sums, key=quarterly_sums.get)
        driest_quarter = min(quarterly_sums, key=quarterly_sums.get)
        
        # Classify based on quarter distribution
        total_rain = sum(quarterly_sums.values())
        if total_rain > 0:
            if quarterly_sums[wettest_quarter] / total_rain > 0.5:
                quarter_type = f'Heavy {wettest_quarter} Dominant'
            elif quarterly_sums['Q4'] > quarterly_sums['Q1'] * 1.5:
                quarter_type = 'Late Year Peaking'
            elif quarterly_sums['Q1'] > quarterly_sums['Q4'] * 1.5:
                quarter_type = 'Early Year Peaking'
            else:
                quarter_type = 'Evenly Distributed'
        else:
            quarter_type = 'No Rainfall'
    else:
        wettest_quarter = 'Unknown'
        driest_quarter = 'Unknown'
        quarter_type = 'Unknown'
    
    quarterly_classification[location_id] = {
        'quarterly_sums': quarterly_sums,
        'wettest_quarter': wettest_quarter,
        'driest_quarter': driest_quarter,
        'quarterly_type': quarter_type
    }

# Seasonal classification
print("\nCreating seasonal rainfall pattern classification...")
seasonal_classification = {}

for location_id in all_features.index:
    location_data = all_features.loc[location_id]
    seasonal_sums = {}
    
    for season in ['DJF', 'MAM', 'JJA', 'SON']:
        sum_col = f'season_{season}_sum'
        if sum_col in all_features.columns:
            seasonal_sums[season] = location_data[sum_col]
        else:
            seasonal_sums[season] = 0
    
    # Find wettest and driest seasons
    if seasonal_sums:
        wettest_season = max(seasonal_sums, key=seasonal_sums.get)
        driest_season = min(seasonal_sums, key=seasonal_sums.get)
        
        # Classify based on season distribution
        total_rain = sum(seasonal_sums.values())
        if total_rain > 0:
            if seasonal_sums[wettest_season] / total_rain > 0.6:
                season_type = f'Strong {wettest_season} Monsoon'
            elif seasonal_sums['DJF'] + seasonal_sums['MAM'] > seasonal_sums['JJA'] + seasonal_sums['SON']:
                season_type = 'Winter-Spring Dominant'
            elif seasonal_sums['JJA'] + seasonal_sums['SON'] > seasonal_sums['DJF'] + seasonal_sums['MAM']:
                season_type = 'Summer-Autumn Dominant'
            else:
                season_type = 'Dual Peak/Even'
        else:
            season_type = 'No Rainfall'
    else:
        wettest_season = 'Unknown'
        driest_season = 'Unknown'
        season_type = 'Unknown'
    
    seasonal_classification[location_id] = {
        'seasonal_sums': seasonal_sums,
        'wettest_season': wettest_season,
        'driest_season': driest_season,
        'seasonal_type': season_type
    }

# Add classifications to main dataframe
print("\nAdding classifications to main dataframe...")

# Convert dictionaries to DataFrames
monthly_class_df = pd.DataFrame.from_dict(monthly_classification, orient='index')
quarterly_class_df = pd.DataFrame.from_dict(quarterly_classification, orient='index')
seasonal_class_df = pd.DataFrame.from_dict(seasonal_classification, orient='index')

# Join with main dataframe
all_features = all_features.join(monthly_class_df, how='left')
all_features = all_features.join(quarterly_class_df, how='left')
all_features = all_features.join(seasonal_class_df, how='left')

print(f"\nTime-specific classifications added successfully!")
print(f"Monthly types distribution:")
print(all_features['monthly_type'].value_counts())
print(f"\nQuarterly types distribution:")
print(all_features['quarterly_type'].value_counts())
print(f"\nSeasonal types distribution:")
print(all_features['seasonal_type'].value_counts())
```
